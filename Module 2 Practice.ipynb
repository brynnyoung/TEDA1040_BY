```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('companies.csv')
```

```python
df.head()
```

```python
sns.set(rc={'figure.figsize':(10, 5)})
sns.countplot(x='state', data=df)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(20, 5)})
sns.barplot(x='Industry', y='job_openings', data=df, estimator=sum, errorbar=None)
```

```python
sns.set(rc={'figure.figsize':(10, 5)})
sns.lineplot(x='founded', y='growth_percentage', data=df)
```

```python
sns.set(rc={'figure.figsize':(10, 5)})
sns.scatterplot(x='employees', y='estimated_revenues', data=df)
```

```python
sns.set(rc={'figure.figsize':(10,5)})
sns.histplot(x='founded', data=df)
```
