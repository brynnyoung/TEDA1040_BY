```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://btech-data-analytics.github.io/bridgerland-technical-college/bookstore.html').text
```

```python
soup = BeautifulSoup(response, 'html.parser')
```

```python
isbn = []
title = []
author = []
language = []
pages = []
topic = []
price = []

for book in soup.find('table').find_all('tr', class_='book'):
    isbn.append(book.find_all('td')[0].text)
    title.append(book.find_all('td')[1].text)
    author.append(book.find_all('td')[2].text)
    language.append(book.find_all('td')[3].text)
    pages.append(book.find_all('td')[4].text)
    topic.append(book.find_all('td')[5].text)
    price.append(book.find_all('td')[6].text)
```

```python
df = pd.DataFrame({
    'isbn': isbn,
    'title': title,
    'author': author,
    'language': language,
    'pages': pages,
    'topic': topic,
    'price': price
})

df.info()
```

Which author has the most books listed at Bob's Bookstore?

```python
df['author'].value_counts()
```

Which topic of books is listed the most frequently?

```python
df['topic'].value_counts()
```

Which topic of books is the most expensive on average?

```python
# Drop the dollar sign from the prices
df['price'] = df['price'].str.replace('$', '', regex=False)
```

```python
# Convert prices to float
df['price'] = df['price'].astype(float)
```

```python
# Calculate average price per topic
df[['topic','price']].groupby('topic').mean().sort_values(by='price', ascending=False)
```

Which topic of book has the most pages, ov average?

```python
# Convert pages to integer
df['pages'] = df['pages'].astype(int)
```

```python
# Calculate average pages per topic
df[['topic', 'pages']].groupby('topic').mean().sort_values(by='pages', ascending=False)
```
