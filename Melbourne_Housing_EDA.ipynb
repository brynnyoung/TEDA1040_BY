## Exploratory Data Analysis on Melbourne Housing

Melbourne, Australia's second-largest city, is known for its vibrant cultural scene, diverse population, and beautiful parks. It boasts iconic landmarks like Federation Square and the Royal Botanic Gardens, as well as a thriving coffee culture. Melbourne is famous for its sports passion, hosting events like the Australian Open and Melbourne Cup, making it a dynamic and cosmopolitan destination.

Recently, the housing market in Melbourne has seen explosive growth. In this project, I will conduct an exploratory data analysis to analyze trends in the housing market and try to determine trends that might be useful to an investor.

I will be using the matplotlib and Seaborn packages to complete the following tasks in order to visualize some key trends:



*   Create a scatter plot that shows land size on the x-axis and price on the y-axis. Change the color of the points to reflect the type of the property.
*   Create a line chart that shows the year built along the x-axis and the average price per year on the y-axis.
*   Create a bar chart that shows the region on the x-axis and the average price of the property on the y-axis.
*   Create a histogram of property prices.
*   Create a horizontally oriented box plot that shows the years that each property was built on the x-axis and the region on the y-axis.




```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('housing.csv')
```

```python
df.head()
```

```python
df.info()
```

### Land Size vs Price Scatter Plot
(Hue by property type)

```python
sns.scatterplot(x='Landsize', y='Price', data=df, hue='Type')
```

### Year Built vs Avg Price Line chart
(I will need to use aggregation and group by to accomplish this)

```python
avg_df = df.groupby('YearBuilt').agg({'Price': 'mean'})
avg_df
```

```python
sns.lineplot(x='YearBuilt', y='Price', data=avg_df)
```

### Region vs Avg Price Bar Chart
(I will need to use group by and aggregation for this)

```python
avg_reg = df.groupby('Regionname').agg({'Price':'mean'})
avg_reg
```

```python
plt.figure(figsize=(10,5))
plt.xticks(rotation=30)
sns.barplot(x='Regionname', y='Price', data=avg_reg)
```

### Property Price Histogram

```python
sns.histplot(x='Price', data=df)
```

### Year Built vs Region Box Plot
(Horizontally oriented)

```python
sns.boxplot(x='YearBuilt', y='Regionname', data=df, orient='h')
```
